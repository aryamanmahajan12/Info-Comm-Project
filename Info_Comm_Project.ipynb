```python
%%manim -ql -v WARNING GraphPlot


class GraphPlot(Scene):
    def construct(self):
        axes = Axes(
            x_range=(0, 3),
            y_range=(0, 3),
            x_length=6,
            y_length=6,
            axis_config={"include_tip": False},
            #x_axis_config={"numbers_to_include": np.arange(0, 4)},
            #y_axis_config={"numbers_to_include": np.arange(0, 4)},
        )
        axes.to_corner(UL)

        graph1 = self.get_graph(lambda x: 1 / x, x_range=(0.225, 5), color=BLUE)
        graph2 = self.get_graph(lambda x: 1 - x, x_range=(-1, 3), color=GREEN)
        graph1.move_to(axes.c2p(1.225, 1.125))
        graph2.move_to(axes.c2p(1, 1))

        x_label = Tex("Memory Size", font_size=20).next_to(axes.x_axis, RIGHT)
        y_label = Tex("Rate", font_size=20).next_to(axes.y_axis, UP)

        graph1_label = MathTex("Global Caching Gain", color=BLUE).scale(0.8).next_to(graph1, direction=UP)
        graph2_label = MathTex("Local Caching Gain", color=GREEN).scale(0.8).next_to(graph2, direction=RIGHT)

        self.play(Create(axes), Create(graph1), Create(graph2))
        self.play(Write(x_label), Write(y_label))
        self.play(Write(graph1_label), Write(graph2_label))

        self.wait()

    def get_graph(self, func, x_range, color):
        graph = FunctionGraph(
            func,
            x_range=x_range,
            color=color,
            use_smoothing=True,
        )
        return graph
```

```python
%%manim -ql -v WARNING CodedCachingModels

from manim import *


class CodedCachingModels(Scene):
    def construct(self):
        F = 3  # Number of files
        N = 4  # Number of users
        k = 2  # Number of files in cache
        f = 2  # Number of bits per file
        m = k  # Number of files to concatenate bits

        server = Rectangle(height=1.5, width=N, color=WHITE, fill_opacity=0, stroke_color=WHITE)
        server_label = Text("SERVER", font_size=30).scale(0.5).move_to(server)
        users = VGroup(*[
            Circle(radius=0.15, color=WHITE, fill_opacity=0, stroke_color=WHITE)
            .shift(((i - (N-1)/2) * 1.5, -1.5, 0))
            for i in range(N)
        ])
        users_labels = VGroup(*[
            Text(f"User {i+1}", font_size=10).next_to(users[i], DOWN)
            for i in range(N)
        ])
        user_caches = VGroup(*[
            Rectangle(height=0.2, width=0.3, color=WHITE, fill_opacity=0, stroke_color=WHITE)
            .next_to(users[i], DOWN, buff=0.5)
            for i in range(N)
        ])
        cache_label = Text("Local Caches:", font_size=14).next_to(user_caches, LEFT, buff=0.5)

        self.play(Write(server), Write(server_label))
        self.wait()
        self.play(FadeIn(users), Write(users_labels))
        self.wait()
        self.play(Write(user_caches), Write(cache_label))
        self.wait()
```

```python
%%manim -ql -v WARNING Entropy

phi_k = r"\phi_k : [2^F]^N \rightarrow 2^{(f \cdot m)}"
psi_d = r"\psi(d_1,...,d_k) : [2^F]^ N \rightarrow [2^{(f \cdot R)}]"
mu_k = r"\mu(d_1,...,d_k),k : [2^{RF}] \times [2^{FM}] \rightarrow [2^F]"
iff = r"if X and Y are independent then,"
h_x_y_exp_ = r"H(X)+H(Y)"


class Entropy(Scene):
    def construct(self):
        text_coded_caching_gain = Text("Coded Caching Gain").scale(0.9).to_edge(UP)
        text_caching_function = Text("Caching Function").scale(0.5).to_edge(DOWN)
        text_encoding_function = Text("Encoding Function").scale(0.5).to_edge(DOWN)
        text_decoding_function = Text("Decoding Function").scale(0.5).to_edge(DOWN)
        eq1 = MathTex(phi_k)
        eq2 = MathTex(psi_d)
        eq3 = MathTex(mu_k)
        eq4 = Text(iff, color=BLUE).shift(2 * UP)
        eq5 = MathTex(h_x_y_exp_, color=BLUE)

        self.add(text_coded_caching_gain)
        self.wait()
        self.play(FadeOut(text_coded_caching_gain))
        self.play(Write(text_caching_function))
        self.wait()
        self.play(FadeOut(text_caching_function))
        self.play(Write(eq1))
        self.wait()
        self.play(FadeOut(eq1))
        self.play(Write(text_encoding_function))
        self.wait()
        self.play(FadeOut(text_encoding_function))
        self.play(Write(eq2))
        self.wait()
        self.play(FadeOut(eq2))
        self.play(Write(text_decoding_function))
        self.wait()
        self.play(FadeOut(text_decoding_function))
        self.play(Write(eq3))
        self.wait()
        self.play(FadeOut(eq3))
```

```python
%%manim -ql -v WARNING CachingAnimation2

class CachingAnimation2(Scene):
    def construct(self):
        # Create heading
        heading_text = Text("Local Caching Gain", color=WHITE).scale(0.8)
        heading_rect = Rectangle(width=heading_text.get_width() + 0.5, height=1, fill_color=BLACK, fill_opacity=0.8)
        heading_group = Group(heading_rect, heading_text)
        heading_group.to_corner(UL)

        # Create cache and source data boxes
        cache_box = Rectangle(width=4, height=1, fill_color=BLUE, fill_opacity=0.8)  # Increase width to 3
        cache_text = Text("Local_Cache", color=WHITE).scale(0.8).next_to(cache_box, DOWN).shift(UP)

        source_box = Rectangle(width=2, height=1, fill_color=BLUE, fill_opacity=0.8)
        source_text = Text("USER", color=RED).scale(0.8).next_to(source_box, DOWN).shift(DOWN)

        self.play(Create(heading_rect), Write(heading_text),Write(cache_box), Write(cache_text),cache_box.animate.shift(UP * 1),cache_text.animate.shift(UP * 1))
        self.play(Write(source_box), Write(source_text))

        # Create data items
        data_items = ["Data 1", "Data 2", "...", "Data k"]
        cache_items = []

        data_objects = []  # Store the mobjects representing the data items

        for i, data_item in enumerate(data_items):
            item_text = Text(data_item, color=WHITE).scale(0.8)
            item_text.move_to(cache_box.get_center() + UP*1.5)  # Adjust position for each item

            # Check if item is in cache
            if data_item in cache_items:
                item_text.set_color(YELLOW)  # Highlight cached items

            data_objects.append(item_text)  # Add item to data objects

            self.play(
                FadeIn(item_text),
                # Move item to the source box
                Transform(item_text, item_text.copy().move_to(source_box.get_center()))
            #self.wait(0.5)
            )
            #self.wait(0.5)
            #if i > 0:
                #fade_out_anims = [FadeOut(obj) for obj in data_objects[:-1]]
                #self.play(*fade_out_anims)
            # Move item to the source box
            self.play(Transform(item_text, item_text.copy().move_to(source_box.get_center())))
            #self.wait(0.5)

            # Move item back to the cache box
            #self.play(Transform(item_text, item_text.copy().move_to(cache_box.get_center())))
            #self.wait(0.5)

            cache_items.append(data_item)  # Add item to cache

            # Apply fade-out effect to all data items except the currently displayed one
            if i >= 0:
                fade_out_anims = [FadeOut(data_objects[i])]
                self.play(*fade_out_anims)

        fade_out_anims = [FadeOut(source_text), FadeOut(*data_objects)]
        self.play(*fade_out_anims)

        source_text = Text("Data", color=WHITE).scale(0.8)

        cache_box_copy1 = cache_box.copy()
        source_box_copy1 = source_box.copy()
        cache_text_copy1 = cache_text.copy()
        source_text_copy1 = source_text.copy()

        cache_box_copy2 = cache_box.copy()
        source_box_copy2 = source_box.copy()
        cache_text_copy2 = cache_text.copy()
        source_text_copy2 = source_text.copy()


        # Minimize the figure and locate it to the bottom-left corner of the scene
        self.play(
            #heading_group.animate.scale(0.5).to_corner(DL),
            cache_box_copy1.animate.scale(0.5).to_corner(DL),
            source_box_copy1.animate.scale(0.5).to_corner(DL),
            cache_text_copy1.animate.scale(0.5).to_corner(DL),
            source_text_copy1.animate.scale(0.5).to_corner(DL),


            cache_box_copy2.animate.scale(0.5).to_corner(DR),
            source_box_copy2.animate.scale(0.5).to_corner(DR),
            cache_text_copy2.animate.scale(0.5).to_corner(DR),
            source_text_copy2.animate.scale(0.5).to_corner(DR),


            cache_box.animate.scale(0.5).to_corner(DOWN),
            source_box.animate.scale(0.5).to_corner(DOWN),
            cache_text.animate.scale(0.5).to_corner(DOWN),
            source_text.animate.scale(0.5).to_corner(DOWN),

        )

        self.play(
            cache_box_copy1.animate.shift(UP * 0.5),
            cache_text_copy1.animate.shift(UP * 0.5),
            cache_box_copy2.animate.shift(UP * 0.5),
            cache_text_copy2.animate.shift(UP * 0.5),
            cache_box.animate.shift(UP * 0.5),
            cache_text.animate.shift(UP * 0.5),
        )


```
